```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Step 1: Prepare Data
x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Independent variable
y = np.array([2, 4, 5, 4, 5])                # Dependent variable

# Step 2: Create and Train Model
model = LinearRegression()
model.fit(x, y)

# Step 3: Predict
y_pred = model.predict(x)

# Step 4: Print model info
print(f"Coefficient (slope): {model.coef_[0]:.4f}")
print(f"Intercept: {model.intercept_:.4f}")
print(f"R-squared: {model.score(x, y):.4f}")

# Step 5: Error Metrics
mae = mean_absolute_error(y, y_pred)
rmse = np.sqrt(mean_squared_error(y, y_pred))  # ✅ Manual RMSE
rae = np.sum(np.abs(y - y_pred)) / np.sum(np.abs(y - np.mean(y)))  # RAE formula

print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"Relative Absolute Error (RAE): {rae:.4f}")

# Step 6: Plot
plt.scatter(x, y, color='blue', label='Actual')
plt.plot(x, y_pred, color='red', label='Predicted')
plt.title('Simple Linear Regression')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Step 1: Prepare Data
x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Independent variable
y = np.array([2, 4, 5, 4, 5])                # Dependent variable

# Step 2: Create and Train Model
model = LinearRegression()
model.fit(x, y)

# Step 3: Predict
y_pred = model.predict(x)

# Step 4: Print model info
print(f"Coefficient (slope): {model.coef_[0]:.4f}")
print(f"Intercept: {model.intercept_:.4f}")
print(f"R-squared: {model.score(x, y):.4f}")

# Step 5: Error Metrics 
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)               # ✅ Added MSE
rmse = np.sqrt(mse)                               # RMSE from MSE
rae = np.sum(np.abs(y - y_pred)) / np.sum(np.abs(y - np.mean(y)))  # RAE

# Print all errors
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")      # ✅ Print MSE
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"Relative Absolute Error (RAE): {rae:.4f}")

# Step 6: Plot
plt.scatter(x, y, color='blue', label='Actual')
plt.plot(x, y_pred, color='red', label='Predicted')
plt.title('Simple Linear Regression')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
